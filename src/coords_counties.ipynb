```python
import numpy as np
import pandas as pd
import string
import csv
import re
```

### Coord Position for current public institutions, municipalities, etc...

```python
municipalities = {}
with open('data/organization.csv', newline='', encoding='utf-8') as f:
    d = csv.reader(f, delimiter=';')
    for i, row in enumerate(d):
        if i <= 0: # Skip header.
            continue
            
        if row[7]:
            m = str(row[3]).replace(' kommune', '')
            m = re.sub("\s+", ' ', m)
            municipalities[str(row[0])] = m.strip()
```

```python
df = pd.read_csv('data/position.csv', delimiter=';', index_col='tailid')
df.iloc[:,3] = 0.0 # Set Z pos from NA to 0.
df.dropna(inplace=True)

# Drop non county related rows.
dropIndexes = []
for k in df.index:
    if str(k) not in municipalities:
        dropIndexes.append(int(k))
        
df.drop(dropIndexes, inplace=True)

# Set the names:
df['name'] = [municipalities[str(key)] for key in df.index]
```
